## **1. Dengan menggunakan Python Requests dan BeautifulSoup cobalah untuk scraping 1 halaman website unair news (https://unair.ac.id/news) dan judul berita yang ada hari ini**

```python
import requests
import csv
from bs4 import BeautifulSoup

response= requests.get('https://unair.ac.id/news/')
rawhtml = response.text
soup = BeautifulSoup(rawhtml, 'html.parser')

with open('news.csv', 'w', newline='') as csvfile:
    news_writer = csv.writer(csvfile)
    news_writer.writerow(['Judul'])

    for i in soup.find_all('h2'):
        # mencari semua text dengan tag <h1> biasa untuk judul
        print(i.get_text())
        news_writer.writerow([i.get_text()])
```

## **2.	Lalu, lakukan crawling unair news dengan mengkombinasikan Python Request, beautifulsoup, dan for loop, untuk mengambil judul berita dari kategori featured news**

```python
import requests
import csv
from bs4 import BeautifulSoup

base_url = 'https://unair.ac.id/en/unair-news-featured/page/{}'
num_pages = 5

with open('featured_news.csv', 'w', newline='') as csvfile:
    news_writer = csv.writer(csvfile)
    news_writer.writerow(['Judul'])

    for page_num in range(1, num_pages+1):
        url = base_url.format(page_num)
        response = requests.get(url)
        rawhtml = response.text
        soup = BeautifulSoup(rawhtml, 'html.parser')

        for i in soup.find_all('h3'):
            title = i.get_text().replace('FEATURED', '')
            print(title)
            news_writer.writerow([title])
```

4. **Membuat code scraping (di VSCode)**

```python
import scrapy


class scrapingspider(scrapy.Spider):
    name = "scrapingspider"
    start_urls = ["https://store.playstation.com/en-id/category/05a2d027-cedc-4ac0-abeb-8fc26fec7180/"]

    def parse(self, response):
        titles = response.css('span.psw-t-body.psw-c-t-1.psw-t-truncate-2.psw-m-b-2::text').getall()
        prices = response.css('span.psw-m-r-3::text').getall()

        for title, price in zip(titles, prices):
            yield {
                'title': title.strip(),
                'price': price.strip()
            }
```
